# **VW05p_Enrichment**

## **Tóm tắt đề bài**

Cho ma trận $a$ có kích thước là $n \times m$ $(3\leq n,m ; n \times m \leq 10^6)$ với $0\leq a_{ij} \leq 10^5$. Tìm ma trận có kích thước $3 \times 3$ mà tổng của ma trận đó nhỏ nhất.

## **Ý tưởng**
***Cách 1:*** Dùng quy hoạch động tổng tiền tố $2$ chiều, ta có thể giải bài toán này với độ phức tạp là $O(n\times m)$ .

***Cách 2*** Dùng vét cạn với tất cả các ma trận con có kích thước là $3\times 3$, ta có thể giải bài toán này với độ phức tạp là $O(n\times m\times 9)$

*=>Ta sẽ sử dụng cách $2$ vì lập trình nó sẽ đơn giản hơn*

## **Source Code**



```python
dx=[ -1,-1,-1,0,0,0,1,1,1 ]
dy=[ -1,0,1,-1,0,1,-1,0,1 ]
#Input
n, m = map(int, input().split())
a = [[int(j) for j in input().split()] for i in range(n)]
#Xu Li
ma=1e18
for i in range(0,n):
  for j in range(0,m):
    c=0
    f=1
    for t in range(0,9):
      x=i+dx[t]
      y=j+dy[t]
      if(x>=0 and x<=n-1 and y>=0 and y<=m-1):
        c+=a[x][y]
      else:
        f=0
        break
    if(f==1):
      ma=min(ma,c)
print(ma)
```

# **VW04p_Checkout**
## **Tóm tắt đề bài**
Cho:

$d:$ quãng đường

$a_1:$ gia tốc tăng

$v_{max}:$ vận tốc tối đa

$t:$ thời gian tối thiểu duy trì vận tốc cao nhất

Tìm $t_{min}$ để hoàn thành quãng đường

## **Ý tưởng**
Chọn $vmax$ lớn nhất có thể hoặc đi đúng $t$ giây với vận tốc vmax và $vmax \leq v$ 

## **Souce Code**

```python
import math
d, a1, a2, v, t = map(float, input().split())
a = -1/(2*a1) - 1/(2*a2)
b = -t
c = d
d1 = (-b+math.sqrt(b*b-4*a*c))/(2*a)
d2 = (-b-math.sqrt(b*b-4*a*c))/(2*a)
vmax = max(d1,d2)
if vmax<v:
    v= vmax
ans = v/(2*a1) + v/(2*a2) +  d/v
print(round(ans,12))
```

# **VW07p_Vitamin**
## **Tóm tắt đề bài**
Cho dãy số nguyên dương $a$ độ dài $n$ $(a_i \leq 1000)$ và số nguyên dương X $(X \leq 10^6)$.

$S_i=a_1+a_2+...+a_i$ ( $i \leq n$)

$S_i=S_n+(i-n) \times a_n$ ( $i>n$)

Có: 
*   Chọn $S_i$ sẽ tốn $i$ đơn vị thời gian
*   Từ lần chọn thứ $2$ trở lên mỗi lần chọn mất thêm $1$ đơn vị thời gian

Hỏi: Xuất ra tồn tại bao nhiêu đơn vị thời gian ít nhất sao cho tổng có thể lập thành $X$. Nếu không thì xuất $-1$.

## **Ý tưởng**
Sử dụng kĩ thuật quy hoạch động

Tạo một mảng cộng dồn $pr[i]=pr[i-1]+a[i]$ để thao tác cho nhanh

*   $d[i]$ là đơn vị thời gian ít nhất để có thể lập thành tổng $i$
*   $d[0]=-1$ và $d[i]=INTMAX$ để ta cực tiểu hóa $d[i]$ và kết quả bài toán sẽ là $d[X]$.
*   Công thức quy hoạch động: $d[i]=min(d[i - pr[n] - k * a[n]] + n + k + 1,d[i - pr[j]] + j + 1)$

## **Source Code**








```python
#input & init data
n, x = map(int, input().split())
a = [0]
a[1:n+1] = map(int, input().split())
pr=[]
pr.append(0)
for i in range(1,n+1):
  pr.append(pr[i-1]+a[i])
#Xu Li
d = [int(1e9) for x in range(x+1)]
d[0] = -1
for i in range(1,x+1):
  t=1e9
  k=(i-pr[n])//a[n]
  if (k>0 and d[i - pr[n] - k * a[n]] != 1e9):
			t = d[i - pr[n] - k * a[n]] + n + k + 1
  for j in range (1,n+1):
    if (i - pr[j] >= 0 and d[i - pr[j]] != 1e9):
      t = min(t, d[i - pr[j]] + j + 1)
  d[i] = min(t, d[i])
if(d[x]==1e9 or d[x]==-1):
  print(-1)
else:
  print(d[x])
```
