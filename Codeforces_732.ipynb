#D: AquaMoon and Chess
Nhận xét: Bài này ban đầu tốn khá nhiều thời gian suy nghĩ ra công thức tính tổ hợp nhưng nhận thấy rằng cứ mỗi 2 quân cờ liền kề thì ta xe chúng như một trái bóng và tính tổ hợp chập. Sử dụng 2 pointers cơ bản và nghịch đảo modulo để cài đặt khá là đơn giản

```python
/*
Nguyen Duc Anh Phuc_Jacke
*/

#include "bits/stdc++.h"

using namespace std;

typedef long long ll;
typedef long double ld;
typedef pair<ll, ll> pll;
typedef pair<ll, pll>plll;

const ll m = 998244353;
vector<ll>inv;
vector<ll>gt;
void modularInverse(ll n)
{
	inv.resize(n + 1);
	gt.resize(n + 1);
	inv[0] = inv[1] = 1;
	gt[0] = gt[1] = 1;

	for (ll i = 2; i <= n; i++)
	{
		inv[i] = inv[m % i] * (m - m / i) % m;
		gt[i] = (gt[i - 1] * i) % m;
	}
}

ll gcdExtended(ll a, ll b, ll* x, ll* y);

ll modInverse(ll a)
{
	ll x, y;
	ll g = gcdExtended(a, m, &x, &y);
	{
		ll res = (x % m + m) % m;
		return res;
	}
}

ll gcdExtended(ll a, ll b, ll* x, ll* y)
{
	if (a == 0)
	{
		*x = 0, *y = 1;
		return b;
	}

	ll x1, y1;
	ll gcd = gcdExtended(b % a, a, &x1, &y1);
	*x = y1 - (b / a) * x1;
	*y = x1;
	return gcd;
}
ll ans(ll n, ll k)
{
	return (((gt[n] * modInverse(gt[k]) % m) * modInverse(gt[n - k]))) % m;
}

int main()
{
	ios_base::sync_with_stdio(false); cin.tie(0); cout.tie(0);
	
	modularInverse(1e5 + 1);

	ll tt;
	cin >> tt;
	while (tt--)
	{
		ll n;
		cin >> n;
		string s;
		cin >> s;
		ll dem = 0;
		ll ball = 0;
		for (ll i = 0; i <= n - 1;)
		{
			if (s[i] == '0')
			{
				dem++;
				i++;
			}
			else
			{
				ll j = i;
				while (j <= n - 1 && s[j] == '1')
					j++;
				j--;
				ball += (j - i + 1) / 2;
				dem += (j - i + 1) / 2;
				i = j + 1;
			}
		}

		if (ball == 0)
			cout << 1 << "\n";
		else
			cout << ans(dem, ball) << "\n";
	}
	return 0;
}
```
